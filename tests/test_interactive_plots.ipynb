```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
def pltsin2(ax):
    x = np.linspace(0,1,100)
    y = np.random.random(size=(100,1))
    ax.clear()
    ax.plot(x,y)
    fig.canvas.draw()
```

```python
fig,ax = plt.subplots(1,1)
pltsin2(ax)
```

```python
fig,ax = plt.subplots(1,1)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_xlim(0,1)
ax.set_ylim(0,1)
for f in range(5):
    pltsin2(ax)
    time.sleep(2)
```
