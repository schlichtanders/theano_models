```python
pandas_dict = None
for datasetname in datasetnames:
    def last_layer_to_dict(l, name, **kwargs):
        l = l[1]
        if l is None:
            return {
                'best_epoch':None,
                'best_test_loss':None,
                'test_error_rate':None,
                'best_val_loss': None,
                'val_error_rate': None, 
                "minus_log_s1" : None,
                "batch_size" : None,
                "opt_identifier" : None,
                "opt_momentum" : None,
                "opt_offset" : None,
                "opt_decay" : None,
                "opt_step_rate" : None,
            }
        return {
            'best_epoch': getattr(l, name + '_best_epoch'),
            'best_test_loss': getattr(l, name + '_best_test_loss'),
            'test_error_rate': getattr(l, name + '_test_error_rate'),
            'best_val_loss': getattr(l, name + '_best_val_loss'),
            'val_error_rate': getattr(l, name + '_val_error_rate'),
            "minus_log_s1" : l.minus_log_s1,
            "batch_size" : l.batch_size,
            "opt_identifier" : l.opt_identifier,
            "opt_momentum" : l.opt_momentum,
            "opt_offset" : l.opt_offset,
            "opt_decay" : l.opt_decay,
            "opt_step_rate" : l.opt_step_rate,
        }
    best_hyper = eva.get_best_hyper(["withpercent"], Hyper, test_suffix=["best_val_loss"], key=lambda fn, p: datasetname in fn)
    pandas_dict = eva.to_pandas_dict(datasetname, best_hyper,
                                     pandas_dict=pandas_dict, last_layer_to_dict=last_layer_to_dict)
```

```python
df = pd.DataFrame(pandas_dict)
df
```

```python
df.to_csv("all_best_hyper_plus_with_params_loss.csv")
# df.to_csv("all_best_hyper_plus_with_params_errorrate.csv")
```

# look at statistics

```python
folder = "evaluation"
datasetname = "boston"
suffix = "_tests.pkl"
```

```python
datasetnames = ["boston","concrete","energy","powerplant","toy1d","toy2d","winered","yacht"]
```

```python
def last_layer_to_dict(l, **kwargs):
    return {
        'best_epoch_mean': l['best_epoch'].mean(),
        'best_epoch_std': l['best_epoch'].std(),
        'best_test_loss_mean': l['best_test_loss'].mean(),
        'best_test_loss_std': l['best_test_loss'].std(),
        'test_error_rate_mean': l['test_error_rate'].mean(),
        'test_error_rate_std': l['test_error_rate'].std(),
    }
```

```python
pandas_dict = None
for datasetname in datasetnames:
    filename = os.path.join(__path__, folder, datasetname + suffix)
    best_tests = evaluation_util.load_dict(filename)
    pandas_dict = evaluation_util.to_pandas_dict(datasetname, best_tests, pandas_dict, last_layer_to_dict=last_layer_to_dict)
```

```python
df = pd.DataFrame(pandas_dict)
```

```python
df.to_csv("all_withpercent_tested_plus.csv")
```
